```python
data = data[ data.Datetime.map(lambda t: start<=t<=finish) ]
data.sort_values('Datetime')
```

```python
print(data.Datetime.min())
print(data.Datetime.max())
```

```python
data['day'] = data.Datetime.map(lambda t:t.strftime('%Y-%m-%d'))
```

```python
day_sum = data.groupby('day').sum()
day_sum
```

```python
day_sum['ds'] = day_sum.index 
```

```python
day_sum.columns = ['y','ds']
day_sum
```

```python
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
# fig, ax = plt.subplots(constrained_layout=True)
xs = [datetime.strptime(d, '%Y-%m-%d').date() for d in day_sum['ds'].values]
plt.figure(figsize=(30,10))
plt.plot(xs,day_sum['y'].values)
plt.show()
```

```python
m = Prophet()
m.fit(day_sum)
```

```python
future = m.make_future_dataframe(periods=31*4+29+30*2)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
forecast
```

```python
fig1 = m.plot(forecast)
```
