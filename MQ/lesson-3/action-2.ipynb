```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('./train.csv')
test_data = pd.read_csv('./test.csv')
```

```python
train_data.info()
```

```python
train_data.describe()
```

```python
train_data.describe(include=['O'])
```

```python
train_data.head()
```

```python
train_data['Age'].fillna(train_data['Age'].mean(),inplace=True)
test_data['Age'].fillna(test_data['Age'].mean(),inplace=True)
```

```python
train_data['Age']
```

```python
train_data['Fare'].fillna(train_data['Fare'].mean(),inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(),inplace=True)
```

```python
train_data['Embarked'].value_counts()
```

```python
train_data['Embarked'].fillna('S', inplace=True)
test_data['Embarked'].fillna('S',inplace=True)
```

```python
train_data
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
train_features = train_data[features]
train_labels = train_data['Survived']
test_features = test_data[features]
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dvec=DictVectorizer(sparse=False)
train_features=dvec.fit_transform(train_features.to_dict(orient='record'))
```

```python
train_features[:5]
```

```python
dvec.feature_names_
```

```python
test_features=dvec.fit_transform(test_features.to_dict(orient='record'))
```

```python
test_features[:5]
```

```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_features,train_labels,test_size=0.25,train_size=0.75)
```

```python
pipeline_optimizer = TPOTClassifier(generations=5, population_size=20, verbosity=3,n_jobs=-1)
```

```python
pipeline_optimizer.fit(X_train, y_train)
```

```python
pipeline_optimizer.score(X_test,y_test)
```

```python
pipeline_optimizer.export('tpot_titanic.py')
```
