```python
import torch
import torch.nn as nn
from torch.nn import functional as F
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
X,y = load_boston(return_X_y=True)
# print(X.shape)
train_X,test_X,train_y,test_y = train_test_split(torch.Tensor(X),torch.Tensor(y),test_size=0.2)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(13,10)
        self.fc2 = nn.Linear(10,20)
        self.fc3 = nn.Linear(20,1)
    
    def forward(self,x):
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        return x
```

```python
model = Net()
mse = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
epochs = 1000
for i in range(epochs) :
    y_pred = model.forward(train_X)
    loss = mse(train_y.view(-1,1),y_pred)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print('---------{}------------'.format(i+1))
    print('loss:',loss)
```

```python
from sklearn.metrics import mean_squared_error
pred = model.forward(torch.Tensor(test_X))
```
