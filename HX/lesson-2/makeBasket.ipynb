```python
import pandas as pd
```

```python
data = pd.read_csv('./MakeBasket.csv',header=None,keep_default_na=False)
```

```python
rows = data.iterrows() 
row = next(rows)
tran = []
while row !='finish' :
    s = set(row[1])
    s.discard('')
    tran.append(s)
    row = next(rows,'finish')
```

```python
print(tran)
```
