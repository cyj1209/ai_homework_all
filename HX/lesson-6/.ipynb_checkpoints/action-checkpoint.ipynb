```python
data.loc[data[data['label']=='male'].index,'label'] = 0
data.loc[data[data['label']=='female'].index,'label'] = 1
```

```python
label = data['label'].values
data.drop(columns='label',inplace=True)
data.head()
```

```python
# 我要的是一个二分类的模型,我的样本为 3168* 20 的样本 我的目的是
# 最简单的是LR模型 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier, Perceptron
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
label = list(label)
```

```python
data_array = data.values
ss = StandardScaler()
data_ss = ss.fit_transform(data_array)
```

```python
train_X,test_X,tran_y,test_y = train_test_split(data_ss,label,test_size=0.15)
```

```python
def train_model(name,model):
    model.fit(train_X,tran_y)
    pre = model.predict(test_X)
    score = accuracy_score(test_y,pre)
    print('{}的分数为：{}'.format(name,score))
```

```python
classifiers = [
    ("SGD", SGDClassifier(max_iter=100)),
    ("ASGD", SGDClassifier(average=True)),
    ("Perceptron", Perceptron()),
    ("Passive-Aggressive I", PassiveAggressiveClassifier(loss='hinge',
                                                         C=1.0, tol=1e-4)),
    ("Passive-Aggressive II", PassiveAggressiveClassifier(loss='squared_hinge',
                                                          C=1.0, tol=1e-4)),
    ("SAG", LogisticRegression(solver='sag', tol=1e-1, C=1.e4 / data.shape[0])),
    ('SVM',SVC(gamma='auto'))
]
```

```python
for name,model in classifiers:
    train_model(name,model)
```
