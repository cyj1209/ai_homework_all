### 参考老师的方法用sk_learn中的方法来解决
  - 在jupyter中如果我实例化两个CountVectorizer对象的话第二个对象在调用get_feature会报错 NotFittedError: Vocabulary not fitted or provided


```python
def cut_words(text):
    text = text.strip().replace(' ','').replace('\n','').replace('\r','')
    # 用空格链接是因为countvectorizer 接收的是一个文本，而且是用空格来切分
    return ' '.join(jieba.cut(text))
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
```

```python
courpus = list(map(cut_words,df['content'].values))
countvectorizer = CountVectorizer(encoding='gb18030',min_df=0.015,stop_words=stopwords,analyzer='word')
countvector = countvectorizer.fit_transform(courpus)
```

## 找出相识度Top10
### 其实我要做的的就是在同一个类别里面求 cosine_similarty 在排序
    - 我需要排除这个类别中的新华社

```python
def find_similar_text(text_index,top=10):
    """
        找出相似度top排行的新闻
        Arg：
            text_index：int 新闻的id
            top ：int default=10 排行的前几个
        Return：
            index_list ：list 排名前面的新闻在df中的index
            cs ： consine——similarity的降序排序
    """
    class_id = k_labels[text_index]
    # 保存的是index 
    class_id_list = np.array(np.where(k_labels == class_id))[0] 
    #我还要删除这个类别中是新华社的 我自己也属于新华社的
    index_list = np.array([i for i in class_id_list if '新华' not in df['source'][i]])  
    #如果某一类都是新华社的，全部都排除了应该直接返回 []
    if index_list == []:
        return [],[]
    #计算cosine_similarty
    cs = [ cosine_similarity(tfidf[i],tfidf[text_index])[0][0] for i in index_list]
    #拿到前10 index 
    index_top = np.argsort(cs)[::-1][:10]
    return index_list[index_top],[cs[i] for i in index_top]
```

```python
text_id = 999
```

```python
def get_index(text_id):
    return df[df['id']==text_id].index[0]
```

```python
index_top,value_top = find_similar_text(get_index(text_id))
```
